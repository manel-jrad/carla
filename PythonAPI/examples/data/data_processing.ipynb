```python
import os
from pandas import read_csv
import pandas as pd
import numpy as np 
from math import sqrt
from numpy import ndarray
from datetime import datetime
from PIL import Image
dataset= read_csv('fichier.csv')
dataset=dataset.iloc[:, 1:]
n=len(dataset)
t=n+1
nombreinitiale=dataset['nombre'][0]
nombrefinal=dataset['nombre'][n-1]
#for j in range(nombreinitiale,nombrefinal) :
    #if j not in (dataset['nombre']):
        
        #os.remove('mm/'+ str(j) + '.png')
for i in range(n):
    vitesse=int(dataset['vitesse'][i])
    steer=dataset['steer'][i]
    nombre=int(dataset['nombre'][i])
    if  vitesse==0:
        dataset=dataset.drop([i])
        os.remove('mm/'+ str(nombre) + '.png')
    elif abs(steer) >= 5:
        p=-steer
        img = Image.open('mm/'+ str(nombre) + '.png')
        img.save('mm/'+ str(int(nombrefinal+1)) + '.png') 
        dataset.loc[t] = [-p,nombrefinal+1,vitesse]
        nombrefinal=nombrefinal+1
        t=t+1
        img1 = img.transpose(Image.FLIP_TOP_BOTTOM)
        img1 = img1.rotate(180)  
        img1.save('mm/'+ str(int(nombrefinal+1)) + '.png')
        
        dataset.loc[t] = [p,nombrefinal+1,vitesse]
        nombrefinal=nombrefinal+1
        t=t+1

dataset.to_csv('fichier2.csv')

        
        
        


```
